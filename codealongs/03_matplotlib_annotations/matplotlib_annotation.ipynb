```python
import duckdb
df_nordic = duckdb.query("""--sql
    WITH nordic AS (
             SELECT 
                state,
                SUM("Total foreign born") as "Total foreigners",
                SUM("Sweden") as Swedes,
                SUM("Norway") as Norwegians,
                SUM("Denmark") as Danes,
                SUM("Finland") as Finnish,
            FROM df 
             GROUP BY state
                         )
            FROM nordic n
            SELECT *, n.swedes + n.norwegians + n.danes + n.finnish AS Nordic
                         ORDER BY nordic DESC
    """).df()
df_nordic
```

## Annotations

```python
fig, ax = nordic_plot()
arrowprops = dict(arrowstyle = "->", linewidth = 2, connectionstyle = "arc3, rad=.3")
ax.annotate(text = "Swedes", xy= (0.8, 35000), arrowprops=arrowprops, xytext = (.8, 40000))
```

```python
#gÃ¶ra detta Ã¥teranvÃ¤ndbart med en function
# hÃ¤r Ã¤r en ganska hÃ¥rdkodad funktion
fig, ax = nordic_plot()
def annotate_nordic(text, xy, x_offset, y_offset):
    arrowprops = dict(arrowstyle = "->", linewidth = 2, connectionstyle = "arc3, rad=.3")
    ax.annotate(
        text = text, 
        xy= xy, 
        arrowprops=arrowprops, 
        xytext = [xy[0] + x_offset, xy[1] + y_offset])
annotate_nordic("Swedes", (0.8,35000), 0.05, 7000)
```

```python
#Mer generaliserad funktion
fig, ax = nordic_plot()

# tuple unpacking of ax.containers that is a list of four BarContainers, which contains 3 Artists each
swede_bars, norwegian_bars, danish_bars, finnish_bars = ax.containers

def annotate_nordic(bar_artist, text, x_offset, y_offset):

    xy = (bar_artist.get_x(), bar_artist.get_height())

    # ternary if statement
    #rad = 0.3 if xy[0] < 0 else 0.3
    
    arrowprops = dict(arrowstyle = "->", linewidth = 2, connectionstyle = f"arc3, rad=0.3")
    ax.annotate(
        text = text, 
        xy=(xy[0] + 0.06, xy[1]), 
        arrowprops=arrowprops, 
        xytext = [xy[0] + x_offset, xy[1] + y_offset])
annotate_nordic(swede_bars[1], "Swedes", 0.05, 7000)
annotate_nordic(norwegian_bars[0], "Norwegians", 0.1, 5000)

# TODO for reader:
# draw other annotations 
annotate_nordic(danish_bars[2], "Danes", 0.08, 5000)
annotate_nordic(finnish_bars[1], "Finnish", 0.08, 5000)
#add labels and title
ax.set_xlabel("States")
ax.set_ylabel("Number of immigrants")
ax.set_title("Immigrants to US states from Nordic countries")
# remove legend 
ax.get_legend().remove()
# save the figure
fig.savefig("figures/nordic_immigration_us.png")
```

```python
swe_ger_vs_total = df[["State", "Total foreign born", "Germany", "Sweden"]]
#swe_ger_vs_total_filtered = swe_ger_vs_total[swe_ger_vs_total['State']=='Alabama']
#swe_ger_vs_total_filtered

swe_ger_vs_total_grouped = swe_ger_vs_total.groupby("State").sum()
swe_ger_vs_total_grouped
```

```python
swe_ger_vs_total_grouped_sorted = swe_ger_vs_total_grouped.sort_values(by="Total foreign born", ascending=False)
swe_ger_vs_total_grouped_sorte_top10 = swe_ger_vs_total_grouped_sorted.iloc[:10]
swe_ger_vs_total_grouped_sorte_top10
```

```python
swe_ger_vs_total_grouped_sorte_top10["share_ger"] = swe_ger_vs_total_grouped_sorte_top10['Germany']/ swe_ger_vs_total_grouped_sorte_top10['Total foreign born'] * 100
swe_ger_vs_total_grouped_sorte_top10["share_swe"] = swe_ger_vs_total_grouped_sorte_top10['Sweden'] / swe_ger_vs_total_grouped_sorte_top10['Total foreign born'] * 100
swe_ger_vs_total_grouped_sorte_top10["share_rest"] = (swe_ger_vs_total_grouped_sorte_top10['Total foreign born'] - swe_ger_vs_total_grouped_sorte_top10['Germany'] - swe_ger_vs_total_grouped_sorte_top10['Sweden']) / swe_ger_vs_total_grouped_sorte_top10['Total foreign born']* 100

#swe_ger_vs_total_grouped_sorte_top10["share_swe"] = swe_ger_vs_total_grouped_sorte_top10["share_swe"].astype(str) + ' %'
#swe_ger_vs_total_grouped_sorte_top10["share_ger"] = swe_ger_vs_total_grouped_sorte_top10["share_ger"].astype(str) + ' %'
#swe_ger_vs_total_grouped_sorte_top10["share_rest"] = swe_ger_vs_total_grouped_sorte_top10["share_rest"].astype(str) + ' %'

swe_ger_vs_total_grouped_sorte_top10 = swe_ger_vs_total_grouped_sorte_top10.drop(["Total foreign born", "Sweden", "Germany"], axis=1)
swe_ger_vs_total_grouped_sorte_top10
```

```python
ax = swe_ger_vs_total_grouped_sorte_top10.plot(kind='bar', stacked=True)
ax.set_ylabel("Percentage")
ax.set_title("Share of total Germany vs Sweden")
arrowprops = dict(arrowstyle = "->", linewidth = 2, connectionstyle = "arc3, rad=.3")
ax.annotate(text = "Swedes", xy= (1, 32), arrowprops=arrowprops, xytext = (.8, 40))
ax.annotate(text = "Germans", xy= (0, 12), arrowprops=arrowprops, xytext = (0, 20))
ax.annotate(text = "The rest", xy= (0, 55), arrowprops=arrowprops, xytext = (0, 45))
ax.get_legend().remove()
fig = ax.get_figure()
fig.savefig("figures/share_of_ger_swe.png")
```
