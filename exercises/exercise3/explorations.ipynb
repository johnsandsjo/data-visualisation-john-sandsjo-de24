```python
#exploration ex3.2
import duckdb
import pandas as pd

def generate_dices(num_dices= 1, throws=10):
    df = pd.DataFrame()
    for num in range(1,num_dices+1):
        df[f"Dice {num}"] = duckdb.query(f"""--sql
            SELECT FLOOR(RANDOM()*6+1)
            FROM generate_series({throws-1})
        """).df().astype(int)
    df.index.name = "Throw"

    return df


df3 = generate_dices(10, throws=3000)
df3
```

```python
#number of dices
len(generate_dices(10,3000).columns[1:])
def df_mean():
    total = 0
    for column in generate_dices(10,3000).columns[1:]:
        total += generate_dices(10,3000)[column].sum()/generate_dices(10,3000)[column].count()
    return total/len(generate_dices(10,3000).columns[1:])
```
