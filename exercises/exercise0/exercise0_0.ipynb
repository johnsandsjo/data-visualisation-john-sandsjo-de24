b) Draw a line chart of quantity for each year. Is there a year that should be skipped?

```python
import duckdb
df_year = duckdb.query(
    """--sql
    SELECT year, SUM(quantity) as Quantity
    FROM df
    GROUP BY year
    ORDER BY year
"""
).df().set_index("Year")
df_year
```

```python
#skipping last year as it is not complete
df_year.iloc[:-1].plot()
```

c) Draw a line chart of average CO2 emissions for same years that as in b)

```python
import duckdb
df_year_co2 = duckdb.query(
    """--sql
    SELECT year, SUM(Avg_CO2) as Avg_CO2
    FROM df
    GROUP BY year
    ORDER BY year
"""
).df().set_index("Year").iloc[:-1]
df_year_co2
```

```python
ax = df_year_co2.plot()
ax.set(ylabel= "Average CO2 emissions", title="Average emissions by year")
```

```python
import duckdb
df_year_month_import = duckdb.query(
    """--sql
    SELECT 
        year || '-' || LPAD(CAST(month as VARCHAR),2,'0') as year_month,
        SUM(import) as tot_import
    FROM df
    GROUP BY year, month
    ORDER BY year, month
"""
).df().set_index("year_month").iloc[:-1]
df_year_month_import
```

e) Draw a line chart of all years and months for average CO2 emissions

```python
import duckdb
df_year_month_co2 = duckdb.query(
    """--sql
    SELECT 
        year || '-' || LPAD(CAST(month as VARCHAR),2,'0') as year_month,
        SUM(Avg_CO2) as co2
    FROM df
    GROUP BY year, month
    ORDER BY year, month
"""
).df().set_index("year_month").iloc[:-1]
df_year_month_co2.plot()
```

```python
import duckdb
df_year_month_e_import = duckdb.query(
    """--sql
    SELECT 
        year || '-' || LPAD(CAST(month as VARCHAR),2,'0') as year_month,
        SUM(Import_Electric) as electric_import
    FROM df
    GROUP BY year, month
    ORDER BY year, month
"""
).df().set_index("year_month").iloc[:-1]
df_year_month_e_import.plot()
```

g) Draw a line chart of average diesel share per year

```python
import duckdb
df_year_month_diesel_share_avg = duckdb.query(
    """--sql
    SELECT 
        year || '-' || LPAD(CAST(month as VARCHAR),2,'0') as year_month,
        AVG(Diesel_Share) as avg_diesel_share
    FROM df
    GROUP BY year, month
    ORDER BY year, month
"""
).df().set_index("year_month").iloc[:-1]
df_year_month_diesel_share_avg.plot()
```

```python
#turn it into a function
def plot_summary(agg_func, metric, summary_stats):
    df_to_plot = duckdb.query(
        f"""--sql
        SELECT 
            year || '-' || LPAD(CAST(month as VARCHAR),2,'0') as year_month,
            {agg_func}({metric}) as {summary_stats}
        FROM df
        GROUP BY year, month
        ORDER BY year, month
    """
    ).df().set_index("year_month").iloc[:-1]
    df_to_plot.plot()
plot_summary("AVG", "Diesel_Share", "Diesel_average")
plot_summary("SUM", "Import_Electric", "electric_import")
```
